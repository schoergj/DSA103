```python
import matplotlib.pyplot as plt
import seaborn as sns
```

### Sequential palettes

Perceptionally uniform: colour difference mapped to difference in value (extreme values might be difficult to plot).

```python
# Seaborn color palettes: "mako", "flare", "crest"; can be reversed by adding "_r"
sns.color_palette("rocket_r", as_cmap=True)
```

```python
# Seaborn cubehelix palettes
sns.cubehelix_palette(as_cmap=True)
```

```python
# Matplotlib palettes: "viridis", "cividis", "magma"
sns.color_palette("viridis", as_cmap=True)
```

```python
# Custom cubehelix palette: 
# start = starting position in color wheel (0-3)
# rot = rotation of colour wheel = change of hues
# dark and light: darkest and lightest colours on the palette
sns.cubehelix_palette(start=.5, rot=-.75, dark=.3, light=0.8, as_cmap=True)
```

### Qualitative color palettes

Qualitative palettes are well-suited to representing categorical data because most of their variation is in the hue component.

```python
# Deafult palette in seaborn:
sns.color_palette()
```

```python
# many other palettes available! Try different ones!
sns.color_palette("colorblind")
```

```python
# circular pattern: evenly spaced hues in a circle, constant brightness and saturation.
# Perception by human eye: different intensity for different colours (RGB issue!)
sns.color_palette("hls", 10)
```

```python
# Create a sequential palette with 6 colors
sns.color_palette("Blues", 8)
```

### Diverging palettes
If extreme values are interesting and the middle is less so (e.g. correlation): Similar to sequential, but with two colours.

```python
# Seaborn palettes: e.g. "vlag", "icefire"
sns.color_palette("vlag", as_cmap=True)
```

```python
# Matplotlib palettes: e.g. "Spectral", "coolwarm"
sns.color_palette("Spectral", as_cmap=True)
```

```python
# Custom diverging palette, using the hus1 system: define two angles on colour wheel and saturation (s)
sns.diverging_palette(20, 150, s=70, as_cmap=True)
```
