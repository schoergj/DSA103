```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- Example dataset: enzyme activity under three conditions ---
np.random.seed(42)
conditions = ["Control", "Treatment A", "Treatment B"]
activity = [100, 130, 135]  # mean activity
errors = [5, 8, 10]         # standard deviation

df = pd.DataFrame({"Condition": conditions, "Activity": activity, "SD": errors})

sns.set_theme(style="whitegrid", context="talk")
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# ----------------------------------------------------------
# BAD BAR CHART

colors = plt.cm.rainbow(np.linspace(0, 1, len(df)))

axes[0].bar(df["Condition"], df["Activity"], color=colors, edgecolor="black", zorder=3)
axes[0].set_title(" Misleading Bar Chart", fontsize=14)
axes[0].set_ylabel("Activity (a.u.)")

# Truncated axis exaggerates differences
axes[0].set_ylim(90, 140)

# Remove grid and add clutter
axes[0].grid(False)
axes[0].set_facecolor("#f0f0f0")

# Random shadows / 3D illusion (optional)
for i, bar in enumerate(axes[0].patches):
    bar.set_linewidth(1.5)
    bar.set_linestyle("--")

# ----------------------------------------------------------
# GOOD BAR CHART

sns.barplot(data=df, x="Condition", y="Activity",
            yerr=df["SD"], capsize=0.2, color="skyblue", ax=axes[1])
axes[1].set_title("Better Bar Chart")
```
