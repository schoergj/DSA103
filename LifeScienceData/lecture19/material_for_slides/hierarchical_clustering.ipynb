Mini-example for hierarchical clustering:

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# --- Create a tiny, easy-to-read dataset ---
# Two obvious clusters:
# Cluster A ~ (1, 1)
# Cluster B ~ (5, 5)

data = np.array([
    [1.0, 1.2],
    [0.8, 0.9],
    [1.1, 0.95],
    [5.0, 5.1],
    [4.8, 4.9],
    [5.2, 5.0]
])

df = pd.DataFrame(data, columns=["Feature_1", "Feature_2"])
df.index = [f"S{i}" for i in range(1, 7)]

# --- Clustermap ---
sns.clustermap(df, cmap="vlag", standard_scale=1, figsize=(6,6))
plt.show()
```
