```python
import pandas as pd
```

### Melt()

```python
df = pd.DataFrame({
    'Measuring point': ['A', 'B', 'C', 'D'],
    'Temperature': [23.2, 25.5, 25.0, 24.1 ],
    'Humidity': [55, 51, 60, 44],
    'Type': ['DHT22', 'DHT11', 'DHT22', 'DHT20']
})
df
```

```python
melted = pd.melt(df, id_vars=['Measuring point', 'Type']) # Specify column names: var_name='', value_name=''
melted
```

### Pivot()

```python
df2 = pd.DataFrame({
    'Measuring point': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Temperature': [23.2, 25.5, 23.4, 25.0, 24.1, 26.2 ],
    'Time': ["03:08:32", "03:08:34", "03:10:54", "03:10:56", "03:12:45", "03:12:50"],
    'Type': ['DHT22', 'DHT11', 'DHT22', 'DHT11', 'DHT22','DHT11']
})
df2
```

```python
pivoted = df2.pivot(index=['Time', 'Type'], columns="Measuring point", values="Temperature")
pivoted
```
